**Variable Length Arguments**

Sometimes we pass variable number of arguments

```python
# def f1(*n): 
```

```python
def sum(*n):
    total=0
    for i in n:
        total=total+i
    print(total)
sum(10,20,30,40)
```

```python
# write the function for finding factorial
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result
factorial(5)
```

```python
# Recursive Method
def factorial(n):
    if n==0:
        result=1
    else:
        result=n*factorial(n-1)
    return result
factorial(6)

    
```

***filter()***

- we can use filter() to filter  values from the given sequence based on some conditions.
- syntax : 
filter(function,sequence)

```python
# To find the even numbers from  the list by using filter()
def isEven(x):
    if x%2==0:
        return True
    else:
        return False
l=[0,5,6,9,60,20,11,42,12,13]
l1=list(filter(isEven,l))
l1
```

```python
# lambda
l=[0,6,9,99,56,34,87]
l1=list(filter(lambda x: x%2==0,l))
l1
```

```python
# odd
l=[0,6,9,99,56,34,87]
l1=list(filter(lambda x:x%2!=0,l))
l1
```

***Map()***

For every element present i the given sequence apply some functionality and generate new element with the required modification for this map()

For every element present in the list perform double and generate new list of double.


 syntax : map(function,sequence)

```python
l=[1,2,3,4,5,6,7]
def doubleit(x):
    return 2*x
l1= list(map(doubleit,l))
l1
```

***reduce function()***

reduce() function reduces sequence of elements into a single element by applying the specified function.

reduce() function present in functools module and hence we should write import statement.

syntax : reduce(function,sequence)

```python
from functools import*
l=[10,20,30,40,50]
result=reduce(lambda x,y:x+y,l)
print(result)
```
