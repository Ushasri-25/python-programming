# Dict Data Structure

If we need to represent the data in the form of **key:"value"** pairs

1. key - unique
2. values can be duplicates

if we need to represent the data in one object other as feauture of object

1. Insertion order is not preserved
2. Duplicates(value) are allowed
3. Mutable
4. Dynamic
5. Indexing and slicing are not applicable

**Creating dictionary**

```python
# blank dictionary
d={}
type(d)
```

```python
# blank dictionary
d=dict()
type(d)
```

```python
d[100]="Anil"
d[200]="pawan"
d[300]="virat"
d
```

```python
d={"key":"pair"}
```

**How to access data from dictionary**

```python
d={100:"abhi",200:"Ishan",300:"Tilak"}
d[100]
```

```python
d[200]
```

**Adding elements**

```python
d[400]="surya"
d
```

```python
# replacing the value
d[300]="vinay"
d
```

**Delete Elements**

```python
d={100:"abhi",200:"Ishan",300:"Tilak"}
del d[200]
d
```

**Functions of dictionary**

***clear()*** - clear the elements from the dictionary

```python
d.clear()
d
```

***len()*** - how many pairs

```python
d=dict()
d=dict({100:"apple",200:"mango"})
d
```

```python
len(d)
```

***get()***

```python
d=dict({100:"apple",200:"mango"})
d.get(100)
```

***pop()*** - To delete the element of a dictionary

```python
d.pop(100)
```

```python
d
```

***popitem()*** - used to delete arbitary item which is in key:vale pair

```python
# It is use to delete arbitary item(key-value)
d={"blue":"sky","sky":"skyblue"}
d.popitem()
```

***keys()*** - Return associated keys with each dictionary

```python
d={1:"10th",2:"12th",3:"degree"}
d.keys()
```

```python
for i in d.keys():
    print(i)
```

***values()*** - to get the values of the dictionary

```python
d={1:"10th",2:"12th",3:"degree"}
d.values()
```

```python
for i in d.values():
    print(i)
```

***copy()***

```python
d1=d.copy()
d1
```

```python
# write a program to take dictionary as input from the keyboard and get the sum of values
d=eval(input("Enter dictionary"))
s=sum(d.values())
print(s)
```

```python
# write a program to take dictionary as an input from the keyboard and get the sum of keys
d=eval(input("Enter key value pairs"))
s=sum(d.keys())
print(s)
```
