# Range data type

- This will represent the a sequence on numbers.
- The elements which will be presented in range  they will be not modifiable.
- Immutable

```python
a=range(10)
```

```python
a
```

```python
for i in a:
    print(i)
```

```python
b=range(0,10)
for i in b:
    print(i)
```

```python
c=range(2,22,2)
for i in c:
    print(i)
```

```python
c[0]
```

```python
# Item assignment is not possible in range
# c[1]=20
```

```python
l=list(range(5,51,5))
print(l)
```

```python
t=tuple(range(3,35,3))
print(t)
```

# Set Data type

- when we want to represent a group of values without duplicate.
- when order is not important then we go for set data type.
1. Insertion order is not preserved 
2. Duplicates are not allowed.
3. Hetrogenuous objects are allowed.
4. index concept is not applicable in set.
5. It is mutable.
6. Growable in nature.
7. we enclosed in {}.

```python
s={10,23.5,5+9j,"Nihan",True,10}
print(s)
```

```python
#s[0]
```

```python
s.add(50)
print(s)
```

```python
s.remove(10)
print(s)
```

# Frozenset Data type

- It is exactly same as set.
- It is immutable.
- add() and remove() will not work here.

```python
print(s)
```

```python
f=frozenset(s)
print(type(f))
```

```python
for i in f:
    print(i)
```
