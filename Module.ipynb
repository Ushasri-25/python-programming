***Random module***

It is use to generate random number.
It is having multiple functions.

Syntax : random() function

```python
from random import *
for i in range(10):
    print(i)
```

```python
from random import *
for i in range(10):
    print(random())
```

***randint()***

To generate random integer between two given numbers

```python
from random import *
for i in range(10):
    print(randint(1,20))
```

***uniform()***

It return random float value between 2 num.

```python
from random import *
for i in range(10):
    print(uniform(1,10))
```

***randrange()***

randrange([start,stop,step])

```python
import random
for i in range(12):
    print(randrange(1,11,2))
```

***choice()***
- It will not return random number
- It will return random object from the given list/tuple

```python
from random import *
list1=["sunny","bunny","funny","chinni","pinni"]
for i in range(10):
    print(choice(list1))
```
