**Functions in list**

**len()** - how many elements present in the list

mylist=[1,2,3,4,5,6,7,8,56,98,6,7,89]
print(len(mylist))

**count()** - how many elements present in the list

```python
m=[1,2,22,2,4,4,55,5,5,5,5,7,8,9]
m.count(2)
```

```python
m.count(5)
```

**index()** - Returns the occurence of first index

```python
m=[1,2,3,4,5,6,1,6,7,7,7,8,9]
m.index(6)
```

**insert()** - Adding an element to a specific index

```python
v=["a","i","o","u"]
v.append("e")
print(v)
```

```python
# To add the element at specific index we can insert the value
v.insert(1,"e")
print(v)
```

**extend()** - to add all items of one list to another

```python
order1=["chicken","mutton","fish","panner"]
order2=["RC","KFC","MCD","Amul"]
order1.extend(order2)
print(order1)
```

**reverse()** - To reverse the order of elements

```python
order1.reverse()
print(order1)
```

**sort()** - To sort the elements (A-Z,a-z,0-9)

```python
N=[0,1,9,4,6,8,3,2,5,7]
sorted=N.sort()
print(N)
```

```python
N=[0,1,9,4,6,8,3,2,5,7]
sorted=N.sort(reverse=True)
print(N)
```

```python
fruits=["Apple","Mango","Banana"]
fruits.sort()
print(fruits)
```

```python
fruits.sort(reverse=True)
print(fruits)
```

```python
fruits.sort(reverse=False)
print(fruits)
```
