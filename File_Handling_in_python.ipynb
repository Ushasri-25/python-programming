# File Handling

There are two types of file
1. Text file ----->abc.txt
2. Binary file ----> image,vedio,audio

***opening file in python***

- f=open(filename,mode)
- mode--> read-r write-w append-a
- r+ read and write
- w+ write and read
- a+ append and read
- f.close()
- f= open(abc.txt)

```python
f=open("abc.txt","w")
print("File name",f.name)
print("File name",f.mode)
print("is it readable",f.readable())
print("is it readable",f.writable())
f.close()
```

***writing text into a file***

```python
f=open("abc.txt","w")
f.write("Hello!\n")
f.write("Good Morning")
print("Successfully written into the file")
f.close()
```
