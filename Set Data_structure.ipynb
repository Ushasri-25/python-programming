# Set

If we want to represent a group of unique elements, then we should use set data type.

***Features of set:***

1. Duplicates are not allowed
2. Insertion order is not preserved
3. We can sort the elememts
4. Indexing & slicing are not allowed
5. Hetrogeneous objects are allowed

```python
s={10,30,40,50,}
print(type(s))
```

```python
# dupliactes are not allowed
s={10,10,10,10,10}
print(s)
```

```python
# list to set
list = [10,20,20,20,30,30,30,40,50]
s = set(list)
s
```

```python
# blank dict to blank set
s={}
print(type(s))
```

```python
# using range
s= set(range(5))
s
```

```python
s=set(s)
type(s)
```

```python
type(s)
```

**Functions of Set**

**add()** - for adding one element only

s= {10,20,30}
s.add(40)
s

**update()**- for adding multiple elements

```python
l=[20,50,60,70]
s.update(l)
s
```

**Copy()**

```python
s={10,20,30,40}
s1 = s.copy()
s1
```

**pop()**

```python
# the elements will we removed randomly as index is not preserved in a set
s1.pop()
```

```python
s1
```

**Remove()**

```python
# the elements will not be removed directly as remove takes the element as an argument to remove.
s1.remove(20)
s1
```

**discard()**

```python
# it'll delete the element if it is present in the list, if not discard() will not show error like remove()
s1.discard(10)
s1
```

**clear()**

```python
s.clear()
s
```

 **Mathematical operators of set**

- union() - combine elements present in both sets
- intersection() - common elements only

**union()**

```python
s1={10,20,30}
s2={40,50,60}
s1.union(s2)
```

```python
s1|s2
```

**Intersection()**

```python
s1={10,20,30}
s2={40,50,60,30}
s1.intersection(s2)
```

```python
s1&s2
```
