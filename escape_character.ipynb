# Escape characters

*In string literals we can use escape character to associate a special meaning*

```python
# \n for new line
da="Data\nAnalytics"
print(da)
```

```python
# \t for one tab distance
GD="Artificial\tIntelligence"
print(GD)
```

```python
#  \b for backspace
a="Mani\bPopulation"
print(a)
```

```python
# \v for vertical tab
b="mani\vPopulation"
print(b)
```

```python
x="Machine\rLearning"
print(x)
```

```python
y="Generative\fAI"
print(y)
```

```python
z="Python\'Programming"
print(z)
```

```python
p="Computer Science\\ AIML"
print(p)
```
