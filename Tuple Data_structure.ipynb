# Tuple

Tuple is similar to list, except it is immutable. Once we create a tuple,we cannot modify it.

***Conditions for a tuple***:

1. Insertion order is preserved
2. Duplicates are allowed
3. Hetrogenous objects are allowed
4. we use () to store elements and seperate them by (,)

**Defining a tuple**

Tuple can be defined with or without ().

```python
t=10,20,30,40
print(t)
```

```python
t=(1,2,3,4,5)
print(t)
```

```python
t=()
print(t)
```

```python
# If you remove , after 10 - it'll become an integer
t=(10)
type(t)
```

```python
t=(10,)
type(t)
```

```python
tu=("sunny","bunny","funny")
print(type(tu))
```

```python
list1=("sunny","bunny","funny")
print(type(list1))
```

```python
t=tuple(list1)
print(type(t))
```

```python
t=tuple(range(2,30,3))
print(t)
```

**Mathematical operator on tuple**

- concatenation (+)
- repetition (*)

```python
# concatination
t1 = 10,20,30
t2 = 40,50,60
t3 = t1+t2
print(t3)
```

```python
# multiplication
t4 = t1*3
print(t4)
```

**Comparision Operators**

```python
t1==t2
```

```python
t1<t2
```

**Few Functions in Tuple**

***len()***

```python
tu=(10,30,4,50,60,50,34,10,10,56,90)
print(len(tu))
```

***Count()***

```python
tu.count(10)
```

***index()***

```python
tu.index(30)
```

***Sorted()***

```python
tu=10,90,50,70,20,40
tu_sorted = sorted(tu)
print(tu_sorted)
```

```python
tuple = "a","d","s","c","e"
tuple_sorted = sorted(tuple)
print(tuple_sorted)
```

```python
# min() and max()
tu_sorted
```

***min() & max()***

```python
min(tu_sorted)
```

```python
max(tu_sorted)
```

```python
# compare() 
t1 = 10,20,30
t2 = 40,50,60
t3 = 10,20,30
t1==t3
```

```python
t2>t1
```

***Tuple Packing***

```python

a=10
b=40
c=60
d=70
t=a,b,c,d
print(t)
```

```python
a=10
b=40
c=60
d=70
t=a,b,c,d
print(type(t))
```
