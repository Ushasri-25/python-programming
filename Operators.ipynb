# 4. Bitwise Operator

- This operator can be used with Integer or bool.
- &,|, ^,<<,>>
# conditions
- Bitwise AND(**&**) -> If both bits are 1, result will be 1 - otherwise 0
- Bitwise OR ( **|**) -> If at least one bit is 1, result will be 1 - otherwise 0
- Bitwise XOR (**^**) -> If bits are different, result will be 1 -otherwise 0
- Bitwise left shift (<<) -> Shift bits to the left
- Bitwise right shift (**>>**) -> Shift bits to the right
