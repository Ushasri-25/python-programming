```python
car_sales_missing_data=pd.read_csv("C:/Users/kusha/Downloads/car-sales-missing-data.csv")
car_sales_missing_data
```

```python
car_sales_missing_data["Odometer"]= car_sales_missing_data["Odometer"].fillna(car_sales_missing_data["Odometer"].mean())
car_sales_missing_data
```

```python
car_sales_missing_data["Colour"]= car_sales_missing_data["Colour"].fillna(car_sales_missing_data["Odometer"].mode())
car_sales_missing_data["Colour"]
```

```python
car_sales_missing_data["Colour"]= car_sales_missing_data["Colour"].fillna(car_sales_missing_data["Odometer"].mode().index[0])
car_sales_missing_data["Colour"]
```
