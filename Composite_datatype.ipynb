# Bytes Data types

```python
b=[10,30,40,60,78]
print(type(b))
```

```python
x=bytes(b)
print(type(x))
```

```python
# Bytes data type is immutable
#x[0]=80
```

```python
for i in b:
    print(i)
```

```python
for i in x:
    print(i)
```

```python
x[0]
```

```python
x[1]
```

# Bytesarray Data type

Its same as bytes data type.
It is mutable

```python
y=[23,60,80,34,68]
ba=bytearray(y)
```

```python
for i in ba:
    print(i, end=", ")
```

```python
for i in ba:
    print(i)
```

```python
ba[0]=100
```

```python
for i in ba:
    print(i)
```
