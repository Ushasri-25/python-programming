# Identifier

- A name in python program is called Identifier.
- t can be the name of variable name ,function name ,Module name or class name.

# Rules to define identifiers in python

1. Only alphabets symbols are allowed(upper & lower case):A-Z , a-z
2. Digits are allowed: 0-9
3. Underscore symbol is allowed
4. $ is not allowed
5. Identifier should start with character not with number
6. Identifiers are case sensitive
7. We can't use reserved words(Keywords) as identifiers
