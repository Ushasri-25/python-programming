# Iterative statements

If we want to execute a group of statement in multiple times then we should go for Iterative statement.

Two Types:
1. For loop
2. While loop

**For Loop**

If we want to execute some action for every element present in some sequence (collection of string) then we should go for for loop.

- sequence - a string or any collection.
- body will be executed on each element present in string/collection.

```python
# Syntax:
for x in sequence:
     body
```

```python
s="virat kohli"
for x in s:
    print(x)
```

```python
list1=[23,"Nihan",6,77,68,78]
for x in list1:
    print(x)
```

```python
# To print character  present in string index wise.
name=input("Enter a name: ")
i=0
for x in name:
    print("Character present at",i,"index is:",x)
    i=i+1
```

```python
# print Hello 10 time
a=range(10)
for i in a:
    print("Hello")
```

```python
# print number from 10 to 1
a=range(10)
for i in a:
    print(10-i)
```

```python
# print odd number from 1 to2 0
a=range(20)
for i in a:
    if i%2!=0:
        print(i)
       

```

```python
# print even number from 1 to 20
a=range(20)
for i in a:
    if i%2==0:
        print(i)
```

```python
a=int(input("Enter a number"))
for i in range(1,a*11,a):
      print(i)
      
```

**while loop**

If we want to execute a group of statemts iteratively until some condition false, then go for while loop.

```python
# syntax:
while condition:
    body
```

```python
# print number from 1 to 10 while loop
x=1
while x<=10:
    print(x)
    x=x+1
```

```python
#display the sum of first n number
n= int(input("Enter a number: "))
sum=0
i=1
while i<=n:
    sum=sum+i
    i=i+1
print("The sum of first",n,"number is",sum)
```

```python
# multiplication table
n=int(input("Enter a number: "))
i=1
while i<=10:
    i=i+1
    print(f"{n}x{i}={n*i}")
       
```

```python
# factorial 
n=int(input("Enter a number: "))
factorial = 1
while n>0:
    factorial = factorial*n
    n = n-1
print(factorial)
```

**Nested Loop**

Loop within the loop

```python
for i in range(4):
    for j in range(4):
        print("i=",i,"j=",j)
```

```python
# multiplication
for i in range(2):
    for j in range(5):
        print(f"{i}x{j}={i*j}")
```

**Transfer statements**

***break***

we can use break statement inside loops to break loop execution based on some condition

```python
for i in range(10):
    if i==7:
        print(" Processing is enough,plz break")
        break
    print(i)
```

```python
# using break in list
cart=[100,200,300,600,700,900]
for item in cart:
    if item>=500:
        print("you get 20 percent discount on this order")
        break
    print(item)
```

***continue***

we can use continue statement to skip current iteration and continue to next.

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# using continue in list
cart=[200,400,600,500,900,1000]
for item in cart:
    if item>=500:
        print(" we can not process this item",item)
        continue
    print(item)
```

***Pass***

- pass is a keyword in python .
- In python if block is required which wont do anything then we can define the empty block with pass keyword.

```python
for i in range(100):
    if i%9==0:
        print(i)
    else:
        pass # to avoid error we can use pass 
```
