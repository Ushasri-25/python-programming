```python
# performance comparison: np.sum() vs python's sum()
```

```python
import time
```

```python
data_list=list(range(1000000))
data_np=np.array(data_list)
```

```python
# python sum time
```

```python
start=time.time()
result1=np.sum(data_np)
end=time.time()
print("python sum:",result1)
print("Time:",end-start)
```

```python
start=time.time()
result2=np.sum(data_np)
end=time.time()
print("python sum:",result2)
print("Time:",end-start)
```
