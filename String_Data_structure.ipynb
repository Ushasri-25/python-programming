**Removing spaces from the strings**


There are 3 methods to remove space

1. ***rstrip()***----> remove space right hand side
2. ***lstrip()***----> remove space left hand side
3. ***strip()***-----> remove space both sides

```python
city=input("Enter your city name: ")
scity=city.strip()
if scity=="Hyderabad":
    print("Hello Hyderabadi.....Adaab")
elif scity=="chennai":
    print("Hello Madrasi.......Vanakaam")
elif scity=="Bengaluru":
    print("Hello  Subhodya")
else:
    print("Enter valid city name")
```

```python
city=input("Enter your city name: ")
scity=city.lstrip()
if scity=="Hyderabad":
    print("Hello Hyderabadi.....Adaab")
elif scity=="chennai":
    print("Hello Madrasi.......Vanakaam")
elif scity=="Bengaluru":
    print("Hello  Subhodya")
else:
    print("Enter valid city name")
```
