**Input Statements**

There are two methods to take input:

1. ***raw_input()*** ->python2
2. ***input()*** ->python3

**input()**

Read the data directly from the keyboard.

```python
x=input("Enter x value")
type(x)
```

Note ->If we write only input, so by default it will take as str

```python
b=input("Enter a number")
c=int(b)
print(c)
type(c)
```

```python
type(b)
```

```python
d=int(input("enter a number"))
type(d)
```

```python
e=float(input("Enter number"))
type(e)
```

```python
# write a program to do sum of 2 i/p values
a=int(input("enter a value"))
b=int(input("enter b value"))
print("Sum:",a+b)
```

**How to take multiple i/p at a time**

```python
a,b=[int(x) for x in input("Enter two values: " ).split()]
print("product",a*b)                       
```

***eval()***

It'll take a string and evaluate the result

```python
x=eval("13+23*263/2")
x
```

```python
# taking expression as i/p
x=eval(input("Enter expression"))
x
```

```python
# takin' list as i/p
l=eval(input("Enter a list"))
type(l)
```

***formated string***

```python
name="Remington"
salary=250000
friend="kimberly"
print("hello {0} your salary is {1} and friend is {2}".format(name,salary,friend))
print("hello {x} your salary is {y} and friend is {z}".format(x=name,y=salary,z=friend))
```
